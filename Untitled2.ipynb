```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "data.csv"
data = pd.read_csv(file_path)

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Set the plot style
sns.set(style="whitegrid")

# List of top commodities to visualize
top_commodities = data['commodity'].value_counts().index[:5]

# Filter data for top commodities
filtered_data = data[data['commodity'].isin(top_commodities)]

# Plotting the trends
plt.figure(figsize=(14, 8))

# Loop through each commodity and plot its trend
for commodity in top_commodities:
    commodity_data = filtered_data[filtered_data['commodity'] == commodity]
    plt.plot(commodity_data['date'], commodity_data['price'], label=commodity)

# Set plot labels and title
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.title('Price Trends of Top Commodities in Maharashtra (2023)')
plt.legend(title='Commodity', loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
