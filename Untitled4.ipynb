```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA

# Load the dataset
data = pd.read_csv('rice.csv')  # Replace with your file path
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)

# Plot the data to visualize
plt.figure(figsize=(14, 6))
plt.plot(data['value'], label='7-Day Moving Average Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('7-Day Moving Average Retail Price of Rice')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Perform Augmented Dickey-Fuller test
adf_test = adfuller(data['value'])
print('ADF Statistic:', adf_test[0])
print('p-value:', adf_test[1])
print('Critical Values:', adf_test[4])
```

```python
# First-order differencing
data['value_diff'] = data['value'].diff()

# Drop missing values after differencing
data.dropna(inplace=True)

# Recheck stationarity
adf_test_diff = adfuller(data['value_diff'])
print('ADF Statistic (Differenced):', adf_test_diff[0])
print('p-value (Differenced):', adf_test_diff[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_acf(data['value_diff'], lags=20, ax=plt.gca())
plt.subplot(122)
plot_pacf(data['value_diff'], lags=20, ax=plt.gca())
plt.show()
```

```python
# Ensure the date column is datetime and set it as the index with a frequency
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)

# Set frequency of the index to daily (or adjust based on your data)
data = data.asfreq('D')

# Differencing the data (if not already done)
data['value_diff'] = data['value'].diff().dropna()

# Recheck stationarity if needed
# adf_test_diff = adfuller(data['value_diff'].dropna())

# Fit the ARIMA model again
model = ARIMA(data['value'], order=(1, 1, 1))  # Adjust p, d, q as needed
model_fit = model.fit()

# Print model summary
print(model_fit.summary())

# Forecast future values
forecast = model_fit.forecast(steps=30)
forecast_index = pd.date_range(start=data.index[-1], periods=30, freq='D')

# Plot the results
plt.figure(figsize=(14, 6))
plt.plot(data['value'], label='Historical Price', color='blue')
plt.plot(forecast_index, forecast, label='Forecasted Price', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('ARIMA Forecast of Retail Price of Rice')
plt.grid(True)
plt.legend()
plt.show()
```
