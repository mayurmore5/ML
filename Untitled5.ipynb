```python
import pandas as pd

# Load the dataset (replace 'your_file.csv' with your actual file path)
data = pd.read_csv('gur.csv')

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Set 'date' as the index
data.set_index('date', inplace=True)

# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Display summary statistics of the data
data_summary = data.describe()
print("Data Summary:\n", data_summary)
```

```python
import matplotlib.pyplot as plt

# Plot the time series data
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['value'], color='blue', label='Gur Price')
plt.title('Daily Retail Price of Gur Over Time')
plt.xlabel('Date')
plt.ylabel('Price ()')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Decompose the time series (assuming 365-day seasonality for yearly patterns)
decomposition = seasonal_decompose(data['value'], model='additive', period=365)

# Plot the decomposed components
decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform the ADF test
adf_test = adfuller(data['value'])
print('ADF Statistic:', adf_test[0])
print('p-value:', adf_test[1])
```

```python
# Differencing the data to make it stationary
data_diff = data['value'].diff().dropna()

# Check again for stationarity
adf_test_diff = adfuller(data_diff)
print('ADF Statistic after differencing:', adf_test_diff[0])
print('p-value after differencing:', adf_test_diff[1])
```

```python
from pmdarima import auto_arima

# Auto-ARIMA to select optimal p, d, q parameters
model = auto_arima(data['value'], seasonal=False, trace=True)

# Train the model
model.fit(data['value'])

# Summary of the model
print(model.summary())
```

```python
# Split the data into training and test sets
train_size = int(len(data) * 0.8)
train, test = data['value'][:train_size], data['value'][train_size:]

# Fit the model on training data
model.fit(train)

# Make predictions
predictions = model.predict(n_periods=len(test))

# Plot predictions against the actual values
plt.plot(test.index, test, label='Actual')
plt.plot(test.index, predictions, label='Predicted')
plt.legend()
plt.show()
```

```python
# Forecast future prices (for the next 30 days, for example)
forecast = model.predict(n_periods=30)

# Plot the forecast
plt.plot(data.index, data['value'], label='Historical')
plt.plot(pd.date_range(data.index[-1], periods=30, freq='D'), forecast, label='Forecast')
plt.legend()
plt.show()
```
